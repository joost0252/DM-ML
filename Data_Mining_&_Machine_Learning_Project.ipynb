```python
# Import required packages
import spacy
from spacy import displacy
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd
import spacy_transformers
```

```python
corpus = pd.read_csv("training_data.csv")
```

```python
# Load French language model
nlp = spacy.load("fr_dep_news_trf")

text_corpus = corpus["sentence"]
text = str(text_corpus) #it needs to be a string for the spacy module 

# spaCy object is used to create a document
my_doc = nlp(text)

my_doc
```

```python
type(my_doc)
```

```python
# Create list of tokens
token_list = []

for token in my_doc:
    token_list.append(token.text)

token_list
```

In the token_list the \n are also represented. I do not know why. Also the number of the row is also presented.

```python
# POS
for word in my_doc:
    print(word.text, word.pos_)
```

Below the code tokenizes into sentences, but since our corpus is all single sentences it does not make any sence for us to use this.

```python
# create list of sentence tokens
sents_list = []

for sent in my_doc.sents:
    sents_list.append(sent.text)

print(sents_list)
```

```python
for chunk in my_doc.noun_chunks:
    print(chunk.text, chunk.root.text, chunk.root.dep_,
          chunk.root.head.text)
```

```python
# Let's visualize this
displacy.render(my_doc, style="dep", jupyter= True, options={'distance': 120})
```

In the following code we remove stopwords. In our project we may have to adjust this as stopwords might be a good indicator for the niveau of the text.

```python
#from spacy.lang.fr.stop_words import STOP_WORDS as fr_stop

# Import stopwords from French language
spacy_stopwords = spacy.lang.fr.stop_words.STOP_WORDS

# Print total number of stopwords
print('Number of stopwords: %d' % len(spacy_stopwords))

# Print 20 stopwords
print('20 stopwords: %s' % list(spacy_stopwords)[:20])
```

```python
# Declare list for filtered sentence
filtered_sent = []

# Filter stopwords
for word in my_doc:
    if word.is_stop == False:
        filtered_sent.append(word.text)

filtered_sent
```

In the following stopwords, punctutation and spaces are removed:

```python
# We can also remove the punctuation

## create empty list to sort the words
filtered_sent2 = []
removed_tokens = []

# Filter stopwords, punctuation and spaces
for word in my_doc:
  if (word.is_stop == True) or (word.is_punct == True) or (word.is_space == True):
    removed_tokens.append(word.text)
  else:
    filtered_sent2.append(word.text)

removed_tokens
```
